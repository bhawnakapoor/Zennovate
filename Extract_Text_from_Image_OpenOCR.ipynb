```python
import os
import cv2
import matplotlib.pyplot as plt
from easyocr import Reader
```

```python
# Path to the folder containing images
folder_path = '/content/drive/MyDrive/HRM Notebooks/images/'

# Iterate through files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.jfif')):  # Check if file is an image
        # Load the image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Convert image to RGB (EasyOCR requires RGB format)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Use EasyOCR to extract text
        result = reader.readtext(image_path)

        # Print the extracted text
        print(f"Text extracted from {filename}:")
        for detection in result:
            text = detection[1]
            print(text)

        # Display the image with detected text
        plt.imshow(image_rgb)
        plt.axis('off')
        plt.show()
```

```python
def extract_text_from_url(image_url):
  """Extracts text from an image URL using EasyOCR.

  Args:
      image_url: The URL of the image.

  Returns:
      A list of extracted text strings for each detected text region.
  """
  text_all = ""
  try:
    response = requests.get(image_url)
    image_bytes = io.BytesIO(response.content)
    image_pil = Image.open(image_bytes)

    # Convert PIL image to OpenCV format (BGR)
    image = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)

    # Convert image to RGB (EasyOCR requires RGB format)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    result = reader.readtext(image_rgb)
    # Print the extracted text
    # print(f"Text extracted from {image_url}:")
    for detection in result:
        text = detection[1]
        text_all += text + ' '
    # Display the image with detected text
    # print(text_all)
    # plt.imshow(image_rgb)
    # plt.axis('off')
    # plt.show()
    return text_all


  except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
    return ''
```
