```python
import json

def extract_category(response):
    start_index = response.find('{')
    end_index = response.find('}') + 1
    json_str = response[start_index:end_index]

    try:
        data = json.loads(json_str)
        category = data.get('Category', '')
    except json.JSONDecodeError:
        category = ''

    return category
```
